```python
cont_cols = ['Users', 'New users', 'Engaged sessions', 'Engagement rate',
       'Engaged sessions per user', 'Average engagement time', 'Event count',
       'Conversions', 'Total revenue']
```

```python
x = df4["Country"]
for i in cont_cols:
        plt.bar(x, df4[i])
        plt.xticks(fontsize = 7, rotation = 90)
        plt.yticks(fontsize = 7)
        plt.xlabel("Country", fontsize = 10)
        plt.ylabel(i,fontsize = 10)
        plt.show()
    
```

Insights:

-->The "Engagement rate" is typically calculated as the percentage of engaged sessions out of total sessions for each country.  Since the numerator (engaged sessions) and denominator (total sessions) are both based on the number of sessions, the contribution of each country to the "Engagement rate" is more evenly distributed. Similarly, "Engaged Sessions per User" is calculated by dividing the total number of engaged sessions by the total number of users in each country.
In contrast, the first set of metrics ("Engaged sessions', 'New Users', 'Engagement Sessions', and 'Users') are absolute counts, not ratios or percentages. The values for these metrics are directly based on the total count of sessions or users in each country, so countries with larger populations or higher levels of activity will naturally have higher counts. It happens the same with "Average Engagement Time".
-->The even contribution of all countries to the engagement rate and engaged sessions per user plots could indicate that users from different countries are exhibiting similar engagement behaviors within their sessions, regardless of the total number of sessions or users in each country.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df4' is your DataFrame with categorical variables

# Create subplots
fig, ax = plt.subplots(3, 3, figsize=(10, 6))
ax = ax.ravel()

# Plot each categorical feature using a box plot
for i, feature in enumerate(cont_cols):
    sns.boxplot(data=df4, x=feature, ax=ax[i])
    ax[i].set_xlabel('value')
    ax[i].set_ylabel(feature)

# Adjust the spacing between the subplots
plt.subplots_adjust(wspace=0.2)

# Display the plot
plt.suptitle("BOX PLOT CATEGORICAL VARIABLE")
plt.show()
```

Insights:

-->From the above Box plot charts we can observe that in most of the attributes like "Users", "New Users", "Engaged Sessions"etc., there is only a single outlier. So from the previous charts we can come to a conclusion that this outlier is India.
    

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df4.corr())
```

Insights:

From the above chart we can identify many highly correlated columns like 'Users', 'New users', 'Engaged sessions', 'Event count', and 'Conversions'. It is better to drop these highly correlated columns by keeping only one representative column.

It is better to work with either 'Engagement rate' or 'Engaged sessions per user' attribute, as they show a moderate correlation and might capture similar engagement aspects.

'Total revenue' column has constant values or missing data. It has no variation, you can consider dropping it from the dataset. Even in case of non profit applications the revenue remains zero.  

```python
import pandas as pd
# Assuming 'data' is your DataFrame containing the dataset
# Compute the correlation matrix
correlation_matrix = df4.corr()

# Create a mask to identify highly correlated columns
high_corr_mask = correlation_matrix.abs() >= 0.95

# Find duplicate columns (columns with high correlation)
duplicated_cols = set()
print("duplica", duplicated_cols)
print("high corr", high_corr_mask.columns)
for col in high_corr_mask.columns:
    if col not in duplicated_cols:
        duplicated_cols.update(set(high_corr_mask.index[high_corr_mask[col]]))

# Drop the duplicated columns
data_cleaned = df4.drop(columns=duplicated_cols)

# Drop 'Total revenue' column if it has NaN values
if 'Total revenue' in data_cleaned.columns and data_cleaned['Total revenue'].isnull().all():
    data_cleaned = data_cleaned.drop(columns='Total revenue')
    
```
